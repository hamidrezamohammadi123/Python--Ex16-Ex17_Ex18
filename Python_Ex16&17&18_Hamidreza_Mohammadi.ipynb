```python
# write code here:
"""
Description of our expectations from the program:

Enhance the MyStringclass that inherits from Python's built-in strclass.basedon below details.

New Features to Implement:
        Find All Occurrences (find_all) :
                    Returns a list of all indices where a given substring is found.
        Count Vowels (count_vowels) : 
                    Returns the number of vowels in the string.
        Is Palindrome (is_palindrome): 
                    Checks if the string is a palindrome.

Modifications to Existing strMethods :
        Modify upper Method:
                    Change the upper method so that it only converts consonants to uppercase , 
                    leaving vowels in lowercase

"""

class MyString(str):
    """Initialization with the same Python string function"""
    def __init__(self, string):
        super().__init__(string)

    def find_all(self, substring):
        """Returns a list of all find_words where a given substring is found"""
        find_words = []
        index = 0
        while True:
            start = self.find(substring, start)
            index=index+1
        return find_words

    def count_vowel(self):
        """Returns the number of vowels in the string"""
        vowels_char = "aeiou"
        cnt = 0
        for _ in self:
            if _.lower() in vowels_char:
                cnt = cnt+1
        return cnt

    def is_palindrome(self):
        """
        Checks if the string is a palindrome
        Words that are reversed do not change, such as level radar
        """
        return self == self[::-1]            #   is it equal to its inverse?

    def upper(self):
        """Change the upper method so that it only converts consonants to uppercase, leaving vowels in lowercase"""
        vowels_char = "aeiou"
        new_str = None
        for char in self:
            if char.lower() not in vowels_char:
                new_str += char.upper()
            else:
                new_str += char.lower()
        return new_str

   
```

```python
# write code here:
"""  
Description of our expectations from the program:
Develop a Python-based university management system to demonstrate your understanding of inheritance,
polymorphism, and abstraction. Create classes for Student,Teacher, and University with specific attributes and methods. 

"""

WHITE   = '\033[97m'
BLUE    = '\033[94m'
RED     = '\033[31m'
YELLOW  = '\033[33m'
GREEN   = '\033[32m'
RESET   = '\033[0m'

#---------------------------------------- class University  ----------------------------- 
class University:
    """ 
    A class representing a university with   
        attributes   : name ,country , global ranking  , students and teachers 
        and  methods : add_student , add_teacher
    """
    def __init__(self, name, country, global_ranking):
        self.name = name
        self.country = country
        self.global_ranking = global_ranking
        self.students  = []
        self.teachers  = []


    def add_student(self, student):
        """Add a student to the university"""
        self.students.append(student)


    def add_teacher(self, teacher):
        """Add a teacher to the university"""
        self.teachers.append(teacher)


    def __str__(self):
        return f"{RED}University Name : {self.name}\n    Country        : {self.country}\n    Global ranking : {self.global_ranking}{RESET}  "
    


#---------------------------------------- class Person     ----------------------------- 
class Person:
    """A class representing a person  that  It is used as a basic class for students and teachers"""
    def __init__(self, name, age):
        self.name = name
        self.age  = age
        
#---------------------------------------- class Student     ----------------------------- 

class Student(Person):
    """A class representing a student with    
         attributes   : name ,age , student_id  
    """

        
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id
       

    def __str__(self):
        return f"{GREEN}   Student ==> Name : {self.name:<20}, Age : {self.age:<15}, Student ID : {self.student_id:<20}{RESET}  "


#---------------------------------------- class Teacher     ----------------------------- 
class Teacher(Person):
    """A class representing a teacher  with 
        attributes   : name ,age , teacher_id
        """
    def __init__(self, name, age, teacher_id):
        super().__init__(name, age)
        self.teacher_id = teacher_id
        
    def __str__(self):
        return f"Teacher    ==> Name : {self.name:<20}, Age : {self.age:<15}, Teacher ID : {self.teacher_id:<20}  "
    

#---------------------------------------- class Course     ----------------------------- 
class Course:
    """A class representing a course with 
         attributes   : name course_id,students , teachers
         and  methods : add_student , add_teacher   
    """
    def __init__(self, name, course_id):
        self.name = name
        self.course_id = course_id
        self.students = []
        self.teacher = Teacher

    def add_student(self, student):
        """Add a student to the course"""
        self.students.append(student)

    
    def add_teacher(self, teacher):
        """Add a teacher to the course"""
        self.teacher=teacher
    
    

    def displays(self):
        print("-------------   Display Course  --------------")
        print(f"{YELLOW}Course Name : {self.name:<20}")
        print(f"Course ID   : {self.course_id:<20}")
        print(self.teacher)
        for st in self.students:
            print(st)
            
    

    
#---------------------------------------- main body     ----------------------------- 
    
if __name__ == "__main__":

    u1=University("sharif", "IRAN", 140)        # create university
    print(u1)
                            
    c1=Course("Python","C1001")                 # create  first Course
    t1=Teacher("Zahra Bakhshande",27,"T1001")   
    c1.add_teacher(t1)                          # First teacher Added
    s1= Student("Hamid reza",24,"S1001")                
    s2= Student("Ali",27,"S1002")
    c1.add_student(s1)                          #ُ The first student of this class was added
    c1.add_student(s2)                          #ُ The second student of this class was added


    c2=Course("C++","C1002")                    # create  second Course
    t2=Teacher("Abass Hamedi",48,"T1002")       # second teacher Added
    c2.add_teacher(t2)
    s3= Student("Mahdis",22,"S1003")            #ُ The first student of this class was added
    s4= Student("Mostafa",23,"S1004")           #ُ The second student of this class was added
    c2.add_student(s3)
    c2.add_student(s4)

    c1.displays()                               # The information of lesson 1 and lesson 2 including 
    c2.displays()                               # teachers and students should be displayed


```
